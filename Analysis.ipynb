```python
# Brass Spectrum_Overall.txt

import pandas as pd
import matplotlib.pyplot as plt

# Simulated reading function (replace with actual file reading)
def read_spectrum_data(filepath):
    """Assume data starts after a specific line and read it."""
    data = []
    with open(filepath, 'r') as file:
        start_reading = False
        for line in file:
            if start_reading:
                parts = line.strip().split(',')
                if parts[0] == '#ENDOFDATA   :':
                    break
                energy, count = float(parts[0]), float(parts[1])
                data.append((energy, count))
            if line.strip() == '#SPECTRUM    : Spectral Data Starts Here':
                start_reading = True
    return pd.DataFrame(data, columns=['Energy (keV)', 'Counts'])


# Specify the path to your file
# Path to your EMSA file
file_path = 'Brass Spectrum_Overall.txt'


# Read the data
spectrum_data = read_spectrum_data(file_path)

# Dictionary of labeled peaks (parse from your OXINSTLABEL data)
labeled_peaks = {
    8.048: 'Cu', 8.905: 'Cu', 0.930: 'Cu', 0.832: 'Cu', 0.811: 'Cu',
    8.639: 'Zn', 9.572: 'Zn', 1.012: 'Zn', 0.906: 'Zn', 0.884: 'Zn',
    10.552: 'Pb', 12.614: 'Pb', 12.623: 'Pb', 11.349: 'Pb', 14.765: 'Pb', 15.101: 'Pb',
    9.185: 'Pb', 2.345: 'Pb', 2.653: 'Pb', 1.823: 'Pb', #0.293: 'Pb', 0.276: 'Pb',
    0.277: 'C', 0.525: 'O',
    2.622: 'Cl',# 2.816: 'Cl', 0.181: 'Cl',
    1.487: 'Al',
    6.404: 'Fe', 7.058: 'Fe', #0.703: 'Fe', 0.615: 'Fe'
}
plt.figure(figsize=(24, 12))
plt.plot(spectrum_data['Energy (keV)'], spectrum_data['Counts'], label='Spectrum', color='blue')


# Annotate labeled peaks
for energy, element in labeled_peaks.items():
    # Find the closest match in spectrum_data to the labeled energy
    idx = (spectrum_data['Energy (keV)'] - energy).abs().idxmin()
    count = spectrum_data.loc[idx, 'Counts']
    plt.scatter([energy], [count], color='red', s=12)  # mark the peak
    plt.text(energy, count, element, fontsize=14, verticalalignment='bottom')

# Customize the plot
plt.title('EDS Spectrum _ Overall',fontsize=20)
plt.xlabel('Energy (keV)',fontsize=16)
plt.ylabel('Intensity (Counts)',fontsize=16)
plt.grid(True)
plt.legend()
plt.show()
```
